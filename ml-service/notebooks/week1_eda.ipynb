```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/creditcard.csv")
```

```python
df.describe()
df.info()
```

```python
print(df['Class'].value_counts())
print(df['Class'].value_counts(normalize='True'))
```

```python
df.groupby('Class')['Amount'].describe()
```

```python
df[df['Class']==1]['Time'].hist(bins=50,alpha=0.5,label='fraud',density=True)
df[df['Class']==0]['Time'].hist(bins=50,alpha=0.5,label='normal',density=True)
plt.legend()
plt.show()
```

```python
fraud=df[df["Class"]==1]
normal=df[df["Class"]==0]
fig,axes=plt.subplots(2,2,figsize=(12,8))
for i,col in enumerate(['V14','V2','V4','Amount']):
    #ennumerate because we want index along with each item
    axes[i//2,i%2].hist(fraud[col],alpha=0.5,label='fraud',bins=50,density=True)
    axes[i//2,i%2].hist(normal[col],alpha=0.5,label='normal',bins=50,density=True)
    axes[i//2,i%2].legend()
    axes[i//2,i%2].set_title(col)
plt.tight_layout()
plt.show()
    
```
