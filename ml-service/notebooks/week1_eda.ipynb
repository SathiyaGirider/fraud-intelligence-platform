```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../creditcard.csv')
print("Loaded successfully")
print(df.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.hist(df[df['Class']==0]['Time'], bins=50, color='blue', alpha=0.7)
ax1.set_title('Normal Transactions - Time')
ax1.set_xlabel('Time (seconds)')

ax2.hist(df[df['Class']==1]['Time'], bins=50, color='red', alpha=0.7)
ax2.set_title('Fraud Transactions - Time')
ax2.set_xlabel('Time (seconds)')

plt.tight_layout()
plt.show()
```
